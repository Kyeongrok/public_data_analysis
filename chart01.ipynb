```python
import matplotlib.pyplot as plt
from matplotlib import rc
rc('font', family='Malgun Gothic')

x = ['1월', '2월', '3월']
y = [10, 20, 30]
z = [100, 200, 300]
c = [400, 200, 0]


plt.subplot(1, 2, 1)
plt.title('수량')
plt.plot(x, z, 'r--')

plt.subplot(1, 2, 2)
plt.bar(x, y, color='g')
plt.title('매출')

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.bar(x, y, color='b')
ax2.plot(x, c, color='r')
```
