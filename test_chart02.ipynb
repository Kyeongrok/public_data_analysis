```python
df['경락일자'] = pd.to_datetime(df['경락일자'], format='%Y%m%d')
df['경락일자']
```

```python
gr = df.set_index('경락일자').groupby(pd.Grouper(freq='m'))
gr.sum()

rc('font', family='Malgun Gothic')
fig, ax1 = plt.subplots()
plt.title('2020.1 ~ 2021.3 거래량과 가격 평균')
ax2 = ax1.twinx()
gr.sum()['거래량'].plot(kind='line', ax=ax1, c='g', label='거래량')
gr.mean()['거래가격'].plot(kind='line', ax=ax2, c='r', label='거래가격평균')
ax1.legend(bbox_to_anchor =(0.65, 1.25), ncol = 2)
ax2.legend(bbox_to_anchor =(1.0, 1.25), ncol = 2)
```
